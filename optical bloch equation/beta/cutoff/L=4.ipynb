```python
import numpy as np



def FT_even(N,cutoff):
    """
    Compute Fourier transform coefficients for even N

    Parameters:
        N: integer, must be even
    Returns:
        coeffs: NumPy array containing computed results
    """
    hbar_c = 197.3269804                          
    a_fm = 197.3269804 / cutoff            
    m = 938.92              
    c = 2 * (np.pi ** 2) * (hbar_c ** 2) / ((N ** 3) * m * (a_fm ** 2))

    coeffs = np.zeros(int(N / 2) + 1)

    for delta_j in range(int(N / 2) + 1):
        S = 0
        for n in range(-int(N / 2) + 1, int(N / 2) + 1):
            S += n ** 2 * np.exp(1j * 2 * np.pi * delta_j * n / N)

        coeffs[delta_j] = c * S.real

    return coeffs

print(FT_even(4,cutoff=170))
```

```python
from quspin.basis import boson_basis_general            
import numpy as np         
from quspin.operators import hamiltonian           

L = 4                     
N_3d = L * L * L

coeffs = FT_even(N=4,cutoff=170)

c_2 = -143.2028868560966
c_3 = 107.232753

s = np.arange(N_3d)                            
x = s % L                          
y = (s // L) % L                   
z = s // (L * L)                   

S_x = x + L * (L - 1 - z) + L * L * y
S_y = z + L*y + L*L*(L-1-x)
S_z = (L-1-y) + L*x + L*L*z

T_x = (x + 1) % L + L * y + L * L * z         
T_y = x + L * ((y + 1) % L) + L * L * z         
T_z = x + L * y + L * L * ((z + 1) % L)         

P_x = (L - x - 1) + L * y + L * L * z        
P_y = x + L * (L - y - 1) + L * L * z        
P_z = x + L * y + L * L * (L - z - 1)        

def Tx(s, i):
    """Coordinate transform for translation by i sites along x"""
    x = s % L
    y = (s // L) % L
    z = s // (L * L)
    return int((x + i) % L + L * y + L * L * z)

def Ty(s, i):
    """Coordinate transform for translation by i sites along y"""
    x = s % L
    y = (s // L) % L
    z = s // (L * L)
    return int(x + L * ((y + i) % L) + L * L * z)

def Tz(s, i):
    """Coordinate transform for translation by i sites along z"""
    x = s % L
    y = (s // L) % L
    z = s // (L * L)
    return int(x + L * y + L * L * ((z + i) % L))

R_1 = y + z * L + x * L * L                        

basis_2 = boson_basis_general(
    N_3d,                        
    Nb=2,                        
    kxblock=(T_x, 0),                          
    kyblock=(T_y, 0),                
    kzblock=(T_z, 0),                
    pxblock=(P_x, 0),                
    pyblock=(P_y, 0),                
    pzblock=(P_z, 0),                
    sxblock=(S_x, 0),                 
    r1block=(R_1, 0)                  
)

Ns = basis_2.Ns          
print(f"Basis-space dimension: {Ns}")

basis_3 = boson_basis_general(
    N_3d,                        
    Nb=3,                        
    kxblock=(T_x, 0),                          
    kyblock=(T_y, 0),                
    kzblock=(T_z, 0),                
    pxblock=(P_x, 0),                
    pyblock=(P_y, 0),                
    pzblock=(P_z, 0),                
    sxblock=(S_x, 0),                 
    r1block=(R_1, 0)                  
)

Ns = basis_3.Ns          
print(f"Basis-space dimension: {Ns}")

basis_4 = boson_basis_general(
    N_3d,                        
    Nb=4,                        
    kxblock=(T_x, 0),                          
    kyblock=(T_y, 0),                
    kzblock=(T_z, 0),                
    pxblock=(P_x, 0),                
    pyblock=(P_y, 0),                
    pzblock=(P_z, 0),                
    sxblock=(S_x, 0),                 
    r1block=(R_1, 0)                  
)

Ns = basis_4.Ns          
print(f"Basis-space dimension: {Ns}")


hopping_max = (
    [[coeffs[-1], i, Tx(i, int(L/2))] for i in range(N_3d)] +       
    [[coeffs[-1], i, Ty(i, int(L/2))] for i in range(N_3d)] +       
    [[coeffs[-1], i, Tz(i, int(L/2))] for i in range(N_3d)]         
)

pot = [[coeffs[0] * 3, i] for i in range(N_3d)]

interact = [[0.5 * c_2, i, i, i, i] for i in range(N_3d)]               

three = [[c_3 / 6, i, i, i, i, i, i] for i in range(N_3d)]               

static = [
    ['+-', hopping_max],                  
    ['n', pot],                    
    ["++--", interact],                       
    ['+++---', three]                           
]
for k in range(1, int(L / 2)):            
        hopping_terms = ([[coeffs[k], i, Tx(i, k)] for i in range(N_3d)] +
                         [[coeffs[k], i, Ty(i, k)] for i in range(N_3d)] +
                         [[coeffs[k], i, Tz(i, k)] for i in range(N_3d)])
        static.append(['+-', hopping_terms])
        static.append(['-+', hopping_terms])

```

```python
H_2 = hamiltonian(
    static, [],
    basis=basis_2,
    dtype=np.float64,
    check_herm=False,           
    check_symm=False,           
    check_pcon=False              
)

H_sparse = H_2.tocsr()
print(f"Number of nonzero Hamiltonian elements: {len(H_sparse.data)}")

E_GS, V_GS = H_2.eigsh(k=1, which='SA')                                   

print(f"Ground-state energy: {E_GS[0]}")
```
