```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------------- 1. text(text) --------------------------
# text(Arialtext,text)
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 12  # text
plt.rcParams['axes.linewidth'] = 1.2  # text
plt.rcParams['figure.dpi'] = 300  # text
plt.rcParams['savefig.dpi'] = 300  # text
plt.rcParams['axes.unicode_minus'] = False  # text

# -------------------------- 2. define5textExamplearray(replace with your real data) --------------------------
# Example:5text,text
arr1 = np.array(energy_2)
arr2 = np.array(energy_3)
arr3 = np.array(energy_4)
arr4 = np.array(energy_5)
arr5 = np.array(energy_6)

# text,text
data_list = [arr1, arr2, arr3, arr4, arr5]
# textlabel(text)
labels = ['L=2', 'L=3', 'L=4', 'L=5', 'L=6']

# -------------------------- 3. text(Nature/Sciencetext) --------------------------
# text:text,text,text,text
#colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']  # text,text,text,text,text(text)
colors = ['#E64B35', '#4DBBD5', '#00A087', '#3C5488', '#F39B7F']
# -------------------------- 4. text --------------------------
fig, ax = plt.subplots(figsize=(8, 6))  # text(text10text,text6text)

# text
for i, arr in enumerate(data_list):
    # textxtext
    x = np.arange(len(arr))
    # text:text,text,text(text),label
    ax.plot(x, arr, color=colors[i], linewidth=3, label=labels[i])

ax.set_yscale('log')  # text:Ytext
ax.tick_params(direction='in')  # text(Scientific plotting standard)
# -------------------------- 5. text(Scientific plotting essentials) --------------------------
# textxtext,ytext(text)
ax.set_xlabel('Iterations', fontsize=28, )
ax.set_ylabel('$\Delta$ $E(MeV)$', fontsize=28, )

# text(text)
#ax.set_title('Comparison of 5 Groups of Arrays', fontsize=16, pad=20)
# text
ax.tick_params(axis='both', labelsize=20, length=3, width=1)
# text(textBest,text)
ax.legend(loc='best', fontsize=24, frameon=True, fancybox=True, shadow=False)

# text(text,text,text)
ax.grid(True, linestyle='--', alpha=0.3, color='gray', linewidth=0.8)

# text:text,text(text)
#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)

# text(optional,text)
ax.set_xlim(0, max(len(arr) for arr in data_list) +50)  # xtext
ax.set_ylim(np.min(np.concatenate(data_list))*0.1, np.max(np.concatenate(data_list)) *5)  # ytext

# text
ax.tick_params(axis='both', labelsize=20, length=3, width=1)

# -------------------------- 6. text(text,text) --------------------------
# textPNG(text,text)textPDF(text,text)
plt.tight_layout()  # text,text
#fig.savefig('scientific_array_plot.png', dpi=300, bbox_inches='tight')  # PNGtext

fig.savefig('/home/lzs/N3L.pdf', bbox_inches='tight',dpi=300)  # PDFtext(text)

# text
plt.show()
```
